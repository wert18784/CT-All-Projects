```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator

from tqdm import tqdm
```

```python
datagen = ImageDataGenerator(
rotation_range=40,
width_shift_range=0.2,
height_shift_range=0.2,
shear_range=0.2,
zoom_range=0.2,
horizontal_flip=True,
fill_mode='nearest')
```

```python
DATADIR = r"F:\Projects\learn\deep learning\convnet\Aug_20"
CATEGORIES = [str(i) for i in os.listdir(DATADIR)]
```

```python
## CODE FOR DATA AUGMENTATION::
for category in tqdm(CATEGORIES):
    path = os.path.join(DATADIR, category)
    class_num = CATEGORIES.index(category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        IMG_SIZE = 75
        img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
        plt.imshow(img_array)
        img_array = img_array.reshape((1,) + img_array.shape)
        
        
        i = 0
        for batch in datagen.flow(img_array, save_to_dir=path, save_prefix='aug', save_format='jpg'):  # this loops runs forever until we break, saving images to current directory with specified prefix
                
            i += 1
            if i > 10:  
                break
        
     
for category in CATEGORIES:
    path = os.path.join(DATADIR, category)
    class_num = CATEGORIES.index(category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        IMG_SIZE = 75
        
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        #plt.imshow(new_array)
```
